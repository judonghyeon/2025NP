````python
class Moneybox:
    def __init__(self, save=0):
        self.save = save  # 저축한 총액을 저장하는 인스턴스 변수

    def deposit(self, money):
        self.save += money
        print(f"저축 후 총액: {self.save}")

    def extract(self, money):
        self.save -= money
        print(f"인출 후 총액: {self.save}")

    def get_total(self):
        return self.save

box = Moneybox()
box.deposit(1000)    # 저축 후 총액: 1000
box.extract(200)     # 인출 후 총액: 800
print(box.get_total())  # 800
````

````python
class Password_manager:
    def __init__(self, initial_password):
        self.old_passwords = [initial_password]  # 과거 패스워드 저장, 마지막이 현재 패스워드

    def get_password(self):
        return self.old_passwords[-1]  # 현재 패스워드 반환

    def set_password(self, new_password):
        if new_password not in self.old_passwords:
            self.old_passwords.append(new_password)
            print("새 패스워드가 설정되었습니다.")
        else:
            print("과거에 사용한 패스워드는 사용할 수 없습니다.")

    def is_correct(self, password):
        return password == self.get_password()

pm = Password_manager("abc123")
print(pm.get_password())         # abc123
pm.set_password("qwerty")        # 새 패스워드가 설정되었습니다.
pm.set_password("abc123")        # 과거에 사용한 패스워드는 사용할 수 없습니다.
print(pm.is_correct("qwerty"))   # True
print(pm.is_correct("wrong"))    # False
````

````python
class Articles:
    def __init__(self, name, stock, price):
        self.name = name      # 제품 이름
        self.stock = stock    # 재고 수량
        self.price = price    # 단가

    def offer_price(self, quantity):
        if 1 <= quantity <= 10:
            return self.price * quantity
        elif 11 <= quantity <= 50:
            return int(self.price * quantity * 0.95)  # 5% 할인
        elif quantity >= 51:
            return int(self.price * quantity * 0.90)  # 10% 할인
        else:
            return 0  # 잘못된 수량

    def purchase(self, quantity):
        if quantity <= self.stock:
            self.stock -= quantity
        else:
            print("재고가 부족합니다.")

item = Articles("mask", 10000, 1000)
price = item.offer_price(40)
print(f"40장 구매 가격: {price}원")  # 5% 할인 적용
item.purchase(40)
print(f"구매 후 재고: {item.stock}장")
````

````python
from abc import ABC, abstractmethod
class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

class Dog(Animal):
    def sound(self):
        return "멍멍"
        
# a = Animal() # 에러: 추상 클래스는 인스턴스화 불가
d = Dog()
print(d.sound()) # "멍멍"

#추상 클래스는 "설계도" 역할만 하며,
#실제 객체는 반드시 모든 추상 메서드를 구현한 자식 클래스에서만 만들 수 있습니다.
````

````python
class Person:
    _count = 1000  # 클래스 변수: ID 시작 번호

    def __init__(self, name, age):
        self.name = name
        self.age = age
        Person._count += 1
        self.__id = Person._count  # 인스턴스 변수: 고유 ID

    def getId(self):
        print(self.__id)

    def getAge(self):
        print(self.age)

class Employee(Person):
    def __init__(self, name, age):
        super().__init__(name, age)

p1 = Person("홍길동", 30)
p2 = Person("김철수", 25)
emp1 = Employee("동양", 65)
emp2 = Employee("서양", 40)

p1.getId()    # 1001
p2.getId()    # 1002
emp1.getId()  # 1003
emp2.getId()  # 1004
````

````python
## 1. 파이썬으로 싱글톤 패턴을 구현하는 방법

### (1) 클래스 변수 활용

class Singleton:
    _instance = None
    def __new__(cls, *args, **kwargs):
        if not cls._instance:
            cls._instance = super().__new__(cls)
        return cls._instance

a = Singleton()
b = Singleton()
print(a is b)  # True

### (2) 데코레이터 활용

def singleton(cls):
    instances = {}
    def getinstance(*args, **kwargs):
        if cls not in instances:
            instances[cls] = cls(*args, **kwargs)
        return instances[cls]
    return getinstance

@singleton
class MyClass:
    pass

a = MyClass()
b = MyClass()
print(a is b)  # True


## 2. 파이썬에서 멀티쓰레드 구현하는 기법

import threading

def worker(num):
    print(f"Worker {num} 실행")

threads = []
for i in range(3):
    t = threading.Thread(target=worker, args=(i,))
    threads.append(t)
    t.start()

for t in threads:
    t.join()


## 3. 멀티쓰레드에서 공유 변수에 대한 Critical Section 보호

### (1) Lock 사용

import threading

counter = 0
lock = threading.Lock()

def increment():
    global counter
    for _ in range(10000):
        with lock:
            # Critical Section
            global counter
            counter += 1

threads = []
for _ in range(5):
    t = threading.Thread(target=increment)
    threads.append(t)
    t.start()

for t in threads:
    t.join()

print(counter)  # 항상 50000

### (2) RLock, Semaphore 등도 사용 가능


#요약
#- 싱글톤: `__new__`, 데코레이터 등으로 구현  
#- 멀티쓰레드: `threading.Thread`  
#- 크리티컬 섹션 보호: `threading.Lock` 등 동기화 도구 사용
````
