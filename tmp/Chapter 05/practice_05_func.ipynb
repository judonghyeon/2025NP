````python
def make_square(n,m):
    for i in range(n):
        for j in range(m):
            print('*', end=' ')
        print()

def each_sum(n):
    total = 0
    for digit in str(n):     
        total += int(digit)   
    return total



def first_diff_index(str1, str2):
    min_len = min(len(str1), len(str2))
    for i in range(min_len):
        if str1[i] != str2[i]:
            return i  
    if len(str1) != len(str2):
        return min_len  
    return -1

def return_num_measure(n):
    measures = []
    for i in range(1, n + 1):
        if n % i == 0:
            measures.append(i)
    return measures

def find_all_indices(s, ch):
    indices = []
    for i in range(len(s)):
        if s[i] == ch:
            indices.append(i)
    return indices

def recursion_sum(n):
    if n == 0:
        return 0
    else:
        return n + recursion_sum(n - 1)



make_square(2, 4)  # print로 감싸지 않음

print(each_sum(123))  # 6
print(each_sum(2025)) # 9

print(first_diff_index("apple", "apricot"))  # 2
print(first_diff_index("hello", "hello"))    # -2
print(first_diff_index("abc", "abcd"))       # 3

print(return_num_measure(12))  # [1, 2, 3, 4, 6, 12]
print(return_num_measure(7))   # [1, 7]

print(find_all_indices("banana", "a"))  # [1, 3, 5]
print(find_all_indices("hello", "l"))   # [2, 3]

print(recursion_sum(100))  # 5050
````

## 함수 실습 정리

### 1. make_square(n, m)
- n행 m열의 별(*) 사각형을 출력하는 함수  
- 반환값이 없으므로 print로 감싸지 않고 호출

```python
make_square(2, 4)
# 출력:
# * * * * 
# * * * * 
```

---

### 2. each_sum(n)
- 숫자 n의 각 자리수를 모두 더해 반환하는 함수

```python
print(each_sum(123))   # 6
print(each_sum(2025))  # 9
```

---

### 3. first_diff_index(str1, str2)
- 두 문자열에서 처음으로 다른 문자의 인덱스를 반환  
- 두 문자열이 완전히 같으면 -2 반환  
- 길이가 다르면 짧은 쪽 이후 인덱스 반환

```python
print(first_diff_index("apple", "apricot"))  # 2
print(first_diff_index("hello", "hello"))    # -1
print(first_diff_index("abc", "abcd"))       # 3
```

---

### 4. return_num_measure(n)
- n의 모든 약수를 리스트로 반환

```python
print(return_num_measure(12))  # [1, 2, 3, 4, 6, 12]
print(return_num_measure(7))   # [1, 7]
```

---

### 5. find_all_indices(s, ch)
- 문자열 s에서 문자 ch가 등장하는 모든 인덱스를 리스트로 반환

```python
print(find_all_indices("banana", "a"))  # [1, 3, 5]
print(find_all_indices("hello", "l"))   # [2, 3]
```

---

### 6. recursion_sum(n)
- 1부터 n까지의 합을 재귀적으로 계산하여 반환

```python
print(recursion_sum(100))  # 5050
```
