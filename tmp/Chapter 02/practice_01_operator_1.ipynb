````python
# 산술 및 관계연산자 연습문제
def add(a,b):
    c = a+b
    return c
print(add(5,3))

def minus(a,b):
    c = a-b
    return c    
print(minus(7,3))

def sub(a,b):
    c = a/b
    return c
print(sub(4,2))

def div(a,b):
    c = a//b
    q = a%b
    return c, q  
print(div(8,3))

def tower(a):
    c = a**5
    return c
print(tower(2))

def compare(a):
    if a>10 and a<20:
        return True
    else:
        return False
print(compare(15))


def string_add():
    a = "Hello"
    b = "World"
    c = a + " " + b
    return c
print(string_add())

def sub_check():
    if 7/2 > 3:
        return True
    else:   
        return False
print(sub_check())

def same_check(a):
    if a==5:
        return True
    else:
        return False
def between_check(a):
    if a>10 and a<20:
        return True
    else:
        return False      
print(same_check(5))
print(between_check(15))  
````

## 산술 및 관계연산자 연습문제 요약

### 1. 기본 산술 연산 함수
- `add(a, b)`: 두 수의 합 반환  
- `minus(a, b)`: 두 수의 차 반환  
- `sub(a, b)`: 두 수의 나눗셈 결과 반환  
- `div(a, b)`: 두 수의 몫과 나머지 반환  
- `tower(a)`: a의 5제곱 반환

#### 예시 코드
```python
def add(a, b):
    return a + b
print(add(5, 3))

def minus(a, b):
    return a - b
print(minus(7, 3))

def sub(a, b):
    return a / b
print(sub(4, 2))

def div(a, b):
    return a // b, a % b
print(div(8, 3))

def tower(a):
    return a ** 5
print(tower(2))
```

---

### 2. 관계 연산 및 논리 연산 함수
- `compare(a)`: a가 10보다 크고 20보다 작으면 True, 아니면 False 반환  
- `same_check(a)`: a가 5와 같으면 True, 아니면 False 반환  
- `between_check(a)`: a가 10보다 크고 20보다 작으면 True, 아니면 False 반환  
- `sub_check()`: 7/2가 3보다 크면 True, 아니면 False 반환

#### 예시 코드
```python
def compare(a):
    return a > 10 and a < 20
print(compare(15))

def same_check(a):
    return a == 5
def between_check(a):
    return a > 10 and a < 20
print(same_check(5))
print(between_check(15))

def sub_check():
    return 7/2 > 3
print(sub_check())
```

---

### 3. 문자열 연산 함수
- `string_add()`: "Hello"와 "World"를 공백으로 연결하여 반환

#### 예시 코드
```python
def string_add():
    a = "Hello"
    b = "World"
    return a + " " + b
print(string_add())
```
