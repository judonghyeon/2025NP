## 연습문제 2 요약

### 1. 0~49까지의 수와 1~50까지의 제곱 리스트 만들기

- `loop_list(list1, list2)` 함수는
  - `list1`에 0부터 49까지의 정수를 추가
  - `list2`에 1부터 50까지의 제곱수를 추가
  - 두 리스트를 출력

#### 코드 예시
```python
list1 = []
list2 = []
def loop_list(list1, list2):
    for i in range(50):
        list1.append(i)
    for i in range(1, 51):
        list2.append(i**2)
    print(list1)
    print(list2)

loop_list(list1, list2)
```

---

### 2. 두 리스트의 같은 인덱스끼리 더해서 새 리스트 만들기

- `sum_list(list3, list4)` 함수는
  - `list3`와 `list4`의 같은 인덱스 요소를 더해 `list5`에 저장
  - 결과 리스트를 출력

#### 코드 예시
```python
list3 = [1, 2, 3]
list4 = [4, 5, 6]
list5 = []
def sum_list(list3, list4):
    for i in range(3):
        list5.append(list3[i] + list4[i])
    print(list5)
print(sum_list(list3, list4))
```

### 3. 사용자로부터 5개의 정수를 입력받아 합과 리스트에 저장

- `sum_of_list(list6, list7)` 함수는
  - 사용자에게 5번 정수를 입력받아 `list6`에 저장
  - 입력된 정수의 합을 출력
  - 합을 `list7`에 추가하고 출력

#### 코드 예시
```python
list6 = []
list7 = []
def sum_of_list(list6, list7):
    for i in range(5):
        num = int(input("정수를 입력하세요: "))
        list6.append(num)
    print("입력된 정수의 합은", sum(list6), "입니다.")
    list7.append(sum(list6))
    print(list7)

sum_of_list(list6, list7)
```

### 핵심 정리

- `append(value)`: 리스트에 값 추가
- `for i in range(n)`: 반복문으로 리스트 채우기
- `리스트[i]`: 인덱스를 이용한 요소 접근 및 연산
- `sum(리스트)`: 리스트의 모든 요소 합 구하기
- 사용자 입력값을 리스트에 저장하고 활용하는
