````python
# 논리 연산자 연습문제

def check_between(a):
    if a>3 and a<7:
        return True
    else:
        return False
print(check_between(5))

def check_over_down(x):
    if x>=10 and x<=20:
        return True
    else:
        return False    
print(check_over_down(7))

def check_five_seven(z):
    if z%5 == 0 or z%7==0:
        return True
    else:
        return False
print(check_five_seven(14))

def plus_odd(z):
    if z > 0 and z%2==0:
        return True
    else:
        return False
print(plus_odd(4))

def check_isString_hello_or_world(a):
    if a == "hello" or a == "world":
        return True
    else:
        return False
print(check_isString_hello_or_world("hello"))
print(check_isString_hello_or_world("python"))
````

## 논리 연산자 연습문제 요약

### 1. 특정 범위 내 값 판별
- `check_between(a)`: a가 3보다 크고 7보다 작으면 True, 아니면 False 반환  
  ```python
  print(check_between(5))  # True
  ```

### 2. 10 이상 20 이하 판별
- `check_over_down(x)`: x가 10 이상 20 이하이면 True, 아니면 False 반환  
  ```python
  print(check_over_down(7))  # False
  ```

### 3. 5 또는 7의 배수 판별
- `check_five_seven(z)`: z가 5 또는 7의 배수면 True, 아니면 False 반환  
  ```python
  print(check_five_seven(14))  # True
  ```

### 4. 양수이면서 짝수 판별
- `plus_odd(z)`: z가 0보다 크고 짝수면 True, 아니면 False 반환  
  ```python
  print(plus_odd(4))  # True
  ```

### 5. 문자열이 "hello" 또는 "world"인지 판별
- `check_isString_hello_or_world(a)`: a가 "hello" 또는 "world"면 True, 아니면 False 반환  
  ```python
  print(check_isString_hello_or_world("hello"))   # True
  print(check_isString_hello_or_world("python"))  # False
  ```

- 반환값으로 True/False를 활용하여 조건 결과를 명확히 표현
