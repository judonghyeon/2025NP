````python
# 연습문제1
friend = []
def frient_insert(friend):
    friend.insert(0, '철수')
    print(friend)
    friend.insert(3, '영희')
    print(friend)
    friend.append('민수')
    print(friend)
print(frient_insert(friend))

num_list = [1,2,3]

def num_insert(num_list):
    num_list[1] = 17              
    print(num_list)
    num_list.extend([4, 5, 6])    
    print(num_list)
    num_list.pop(0)               
    num_list.sort()               
    print(num_list)
    num_list.insert(3, 25)       
    print(num_list)

num_list = [1, 2, 3]

print(num_insert(num_list))
````

### 1. 친구 리스트(friend) 다루기

- 빈 리스트 `friend`를 만든다.
- `frient_insert(friend)` 함수는 다음을 수행한다:
  1. 리스트 맨 앞에 '철수'를 삽입한다.
  2. 인덱스 3에 '영희'를 삽입한다(리스트 길이가 3 이상일 때).
  3. 리스트 맨 뒤에 '민수'를 추가한다.
- 각 단계마다 리스트 상태를 출력한다.

#### 코드 예시
```python
friend = []
def frient_insert(friend):
    friend.insert(0, '철수')
    print(friend)
    friend.insert(3, '영희')
    print(friend)
    friend.append('민수')
    print(friend)
print(frient_insert(friend))
```

### 2. 숫자 리스트(num_list) 다루기

- 리스트 `num_list = [1, 2, 3]`을 만든다.
- `num_insert(num_list)` 함수는 다음을 수행한다:
  1. 두 번째 요소를 17로 수정한다.
  2. 리스트에 4, 5, 6을 추가한다.
  3. 첫 번째 요소를 제거한다.
  4. 리스트를 오름차순으로 정렬한다.
  5. 인덱스 3에 25를 삽입한다.
- 각 단계마다 리스트 상태를 출력한다.

#### 코드 예시
```python
num_list = [1, 2, 3]
def num_insert(num_list):
    num_list[1] = 17
    print(num_list)
    num_list.extend([4, 5, 6])
    print(num_list)
    num_list.pop(0)
    num_list.sort()
    print(num_list)
    num_list.insert(3, 25)
    print(num_list)

print(num_insert(num_list))
```

### 핵심 정리

- `insert(index, value)`: 리스트의 원하는 위치에 값 삽입
- `append(value)`: 리스트 맨 뒤에 값 추가
- `extend([values])`: 여러 값을 한 번에 리스트에 추가
- `pop(index)`: 특정 인덱스의 값 제거
- `sort()`: 리스트 정렬
- 리스트의 요소는 인덱스를
