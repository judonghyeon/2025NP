````python
# 멤버 연산자 연습문제
e = [1,2,3,4,5]
def seven_in_list(a):
    if 7 in a:
        return True
    else:
        return False

set = {"orange", "banana", "kiwi"}
def apple_in_set(a):
    if "apple" in a:
        return True
    else:
        return False
    
def is_number_in_dic():
    set = {"a" :1 , "b":2, "c":3}
    if 6 in set.values():
        return True
    else:
        return False
    
def is_in_tuple(b):
    tuple = ("dog", "cat", "bird", "hamster")
    if b in tuple:
        return True
    else:
        return False
    
def is_in_set(a):
    set = {2,4,6,8,10}
    if a in set:
        return True
    else:
        return False


print(seven_in_list(e))
print(apple_in_set(set))
print(is_number_in_dic())
print(is_in_tuple("cat"))
print(is_in_set(10))
````

## 멤버 연산자 연습문제 요약

### 1. 리스트에 특정 값이 있는지 확인
- `seven_in_list(a)`: 리스트 `a`에 7이 있으면 True, 없으면 False 반환  
  ```python
  print(seven_in_list([1,2,3,4,5]))  # False
  ```

### 2. 집합(set)에 특정 값이 있는지 확인
- `apple_in_set(a)`: 집합 `a`에 "apple"이 있으면 True, 없으면 False 반환  
  ```python
  print(apple_in_set({"orange", "banana", "kiwi"}))  # False
  ```

### 3. 딕셔너리 value에 특정 값이 있는지 확인
- `is_number_in_dic()`: 딕셔너리 value에 6이 있으면 True, 없으면 False 반환  
  ```python
  print(is_number_in_dic())  # False
  ```

### 4. 튜플에 특정 값이 있는지 확인
- `is_in_tuple(b)`: 튜플에 `b`가 있으면 True, 없으면 False 반환  
  ```python
  print(is_in_tuple("cat"))  # True
  ```

### 5. 집합(set)에 특정 값이 있는지 확인
- `is_in_set(a)`: 집합에 `a`가 있으면 True, 없으면 False 반환  
  ```python
  print(is_in_set(10))  # True
  ```

### 핵심 정리

- `in` 연산자를 사용하면 리스트, 집합, 딕셔너리, 튜플 등에서 값의 포함 여부를 쉽게 확인할 수 있다.
- 멤버 연산자는 다양한 자료구조에서 조건문과 함께 자주 사용된다.
