````python
# 연습문제-1

days = {'January': 31, 'February': 28, 'March': 31, 'April': 30,
        'May': 31, 'June': 30, 'July': 31, 'August': 31,
        'September': 30, 'October': 31, 'November': 30, 'December': 31}

month = input("월을 입력하세요: ")
if month in days:
    print(f"{month} 는 {days[month]} 일 입니다.")

else:
    print("Invalid month name.")

for month in sorted(days):
    print(f"{month}: {days[month]}일")

for month in days:
    if days[month] == 31:
        print(f"{month}: {days[month]}일")

for month, day in sorted(days.items(), key=lambda x: x[1]):
    print(f"{month}: {day}일")

month_input = input("월을 앞 3글자(영어)로 입력하세요: ").capitalize()
found = False
for month in days:
    if month.startswith(month_input):
        print(f"{month} 는 {days[month]} 일 입니다.")
        found = True
        break
if not found:
    print("Invalid month name.")
````

## 연습문제-1 요약

### 1. 월별 일수 딕셔너리 생성
- `days` 딕셔너리에 각 월의 이름(영어)과 일수를 저장함.

### 2. 사용자 입력으로 월의 일수 출력
- 사용자가 월 이름을 입력하면 해당 월의 일수를 출력
- 딕셔너리에 없는 월을 입력하면 "Invalid month name." 출력

### 3. 월 이름을 알파벳 순서로 출력
- `sorted(days)`를 사용하여 월 이름을 알파벳 순서로 정렬 후 출력

### 4. 일수가 31일인 월만 출력
- 반복문과 조건문을 사용하여 일수가 31인 월만 출력

### 5. 월의 일수를 기준으로 오름차순 출력
- `sorted(days.items(), key=lambda x: x[1])`로 일수 기준 정렬 후 (월, 일수) 쌍 출력

### 6. 월 이름 앞 3글자만 입력해도 일수 출력
- 사용자가 월 이름의 앞 3글자만 입력해도 해당 월의 일수를 찾아 출력
- 해당 월이 없으면 "Invalid month name." 출력
