````python
# 연습문제 1
def mul(a,b):
    c = a*b
    return c
a=int(input())
b=int(input())
print(mul(a,b))

def mileCal(a):
    mile = a * 1.609
    return mile

c = int(input())
print(mileCal(c), "마일")

PI = 3.141592
def CircleExtent(radius):
    CircleEx = PI * radius * radius
    return CircleEx

def CircleArea(radius):
    CircleAr = 2 * PI * radius
    return CircleAr
d = int(input("반지름을 입력하세요: "))

print("반지름이", d, "인 원의 넓이:", CircleExtent(d))
print("반지름이", d, "인 원의 둘레:", CircleArea(d))
````

## 연습문제 1 요약

### 1. 두 수의 곱 구하기
- `mul(a, b)` 함수: 두 수를 곱해서 반환
- 사용자로부터 두 정수를 입력받아 곱셈 결과 출력

#### 예시 코드
```python
def mul(a, b):
    c = a * b
    return c
a = int(input())
b = int(input())
print(mul(a, b))
```

---

### 2. 마일을 킬로미터로 변환
- `mileCal(a)` 함수: 마일 값을 받아 1.609를 곱해 km로 변환
- 사용자로부터 마일 값을 입력받아 변환 결과 출력

#### 예시 코드
```python
def mileCal(a):
    mile = a * 1.609
    return mile

c = int(input())
print(mileCal(c), "마일")
```

---

### 3. 원의 넓이와 둘레 구하기
- `CircleExtent(radius)` 함수: 반지름을 받아 원의 넓이(πr²) 반환
- `CircleArea(radius)` 함수: 반지름을 받아 원의 둘레(2πr) 반환
- 사용자로부터 반지름을 입력받아 넓이와 둘레 출력

#### 예시 코드
```python
PI = 3.141592
def CircleExtent(radius):
    CircleEx = PI * radius * radius
    return CircleEx

def CircleArea(radius):
    CircleAr = 2 * PI * radius
    return CircleAr

d = int(input("반지름을 입력하세요: "))
print("반지름이", d, "인 원의 넓이:", CircleExtent(d))
print("반지름이", d, "인 원의 둘레:", CircleArea(d))
```
