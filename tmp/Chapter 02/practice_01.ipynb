## 연습문제 2 요약

### 1. 초를 분과 초로 변환
- `second_to_mins_sec(a)`: 입력받은 초(`a`)를 분과 초로 변환하여 반환
- 예시: 130초 → (2, 10) (2분 10초)

```python
def second_to_mins_sec(a):
    mins = a // 60
    sec = a % 60
    return mins, sec
a = int(input())
print(second_to_mins_sec(a), "분", "초")
```

---

### 2. 분을 일, 시간, 분으로 변환
- `min_to_day_hour_mins(a)`: 입력받은 분(`a`)을 일, 시간, 분으로 변환하여 반환
- 예시: 1500분 → (1, 1, 0) (1일 1시간 0분)

```python
def min_to_day_hour_mins(a):
    days = a // 1440
    hours = (a % 1440) // 60
    mins = a % 60
    return days, hours, mins
a = int(input())
print(min_to_day_hour_mins(a), "일", "시간", "분")
```

### 3. 복리 계산
- `compound_interest(principal, rate, years)`: 원금, 이율, 기간을 받아 복리 계산 결과 반환
- 예시: 5000000원, 5%, 5년 → 6381401.25

```python
def compound_interest(principal, rate, years):
    total = principal * (1 + rate) ** years
    return total

print(compound_interest(5000000, 0.05, 5))
```

### 4. 1부터 n까지의 합 구하기
- `sum_between(b)`: 1부터 b까지의 합을 반환
- 예시: 100 → 5050

```python
def sum_between(b):
```
